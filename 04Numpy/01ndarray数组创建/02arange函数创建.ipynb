```python
"""
    numpy.arange([start,] stop[, step,], dtype=None)
"""

import numpy as np

# 创建一个0-9的数组
print(np.arange(10))
```

```python
# 创建一个5-14的数组,步长为2的数组
print(np.arange(5,15,2))
```

```python
# 创建一个3-5的数组,步长为0.2的数组
print(np.arange(3,5,0.2))
```

```python

print(np.arange(10,dtype=np.float16))
```
