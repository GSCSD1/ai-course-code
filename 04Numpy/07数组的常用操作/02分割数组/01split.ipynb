```python
"""
    split:用于沿着指定的轴分割成多个数组
    numpy.split(ary, indices_or_sections, axis=0)
    ary: 要分割的数组
    indices_or_sections：
                        可以是一个整数，表示要将数组平分成多少个子数组
                        也可以是一个整数数组，表示分割的位置
    axis：沿着哪个轴进行分割
"""
import numpy as np


# 1维 等分
arr1 = np.arange(10)
print(arr1.shape)
print(arr1)

print(np.split(arr1,2))

# 1维 索引位置分割
print(np.split(arr1,[3,6]))  #arr1[0:3],arr1[3:6],arr1[6:]
print(np.split(arr1,[3,4,6])) # arr1[0:3],arr1[3:4],arr1[4:6],arr1[6:]
```

```python
# 2维 行方向分割
# 2等分
arr = np.random.randint(10,size=(4,3))
print("原数组:\n", arr)
print("分割结果\n",np.split(arr,4,axis=0))

# 2维 索引位置分割
print("分割结果\n",np.split(arr,[1,2],axis=0)) # arr[0:1],arr[1:2],arr[2:3],arr[3:]
```

```python
# 2维 列方向分割
# 3等分
arr = np.random.randint(10,size=(3,6))
print("原数组:\n", arr)
print("分割结果\n",np.split(arr,2,axis=1))

# 2维 索引位置分割
print("分割结果\n",np.split(arr,[2,4],axis=1)) # arr[0:1],arr[1:2],arr[2:3],arr[3:]
```

```python
# 3维
arr = np.random.randint(10,size=(2,3,6))
print("原数组:\n", arr)

print("分割结果\n",np.split(arr,2,axis=0))
```

```python
# 4维
arr = np.random.randint(10,size=(2,2,3,6))
print("原数组:\n", arr)

print("分割结果\n",np.split(arr,2,axis=3))
```
