```python
"""
    where : 可以根据条件返回满足条件的元素的索引
    numpy.where(condition)
"""
import numpy as np

# 1、根据条件返回元素索引
arr = np.array([1,5,8,10,78,10,98,20,30])

print(np.where(arr<30))
```

```python
# 2、根据条件选择元素
# 如果元素大于20，返回'大于20'，否则返回'小于等于20'
arr = np.array([1,5,8,10,78,10,98,20,30])

print(np.where(arr>20,'大于20','小于等于20'))

# 将大于40的值全部改为100
print(np.where(arr>40,100,arr))
```
