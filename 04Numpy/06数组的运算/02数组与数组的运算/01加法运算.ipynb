```python
"""
    数组加法： 各个数组对应的元素相加
"""

import numpy as np

arr1 =  np.random.randint(5,20,size=(2,3))
arr2 =  np.random.randint(5,20,size=(2,3))
print(arr1)
print(arr2)

print(arr1 + arr2)
```

```python
"""
    numpy.add(x1, x2, /, out=None, *, where=True, casting='same_kind',
            order='K', dtype=None, subok=True)

    x1: 第一个输入数组或标量。
    x2: 第二个输入数组或标量。
    out:输出数组
    where：表示计算加法的条件
"""

arr1 =  np.random.randint(5,20,size=(2,3))
arr2 =  np.random.randint(5,20,size=(2,3))
print(arr1)
print(arr2)

print(np.add(arr1,arr2))
```

```python
# where: 表示计算加法的条件，如果设置为True，则在相应位置进行计算；如果 设置为False，则在相应位置不进行计算。
arr1 =  np.random.randint(5,20,size=(2,3))
arr2 =  np.random.randint(5,20,size=(2,3))
print(arr1)
print(arr2)
print(arr1>12)
# print(np.add(arr1,arr2,where=arr1>10))
print(np.add(arr1,arr2,where=np.array([[False,True,False],[True,False,True]])))
```

```python
# out: 输出数组，数据类型必须与预期输出相符。如果提供，它必须具有与输出 相同形状的适当类型来接收结果。
arr1 =  np.random.randint(5,20,size=(2,3))
arr2 =  np.random.randint(5,20,size=(2,3))
print(arr1)
print(arr2)

# 创建一个输出数组，存储结果
arr3 = np.ones_like(arr1)  # 创建一个和arr1形状相同的零数组
print(arr3)
# arr3 = np.add(arr1,arr2)
np.add(arr1,arr2,out=arr3)
print(arr3)
```
