```python
"""
    randn(d0, d1, ..., dn)
    randn:从标准的正态分布中抽取样本，返回的随机数将具有平均值（a）为0，标准差(d)为1的正态分布  样本量较小时还会出现一些看似 “偏离” 的情况

    根据 68-95-99.7规则

    68%的数据会落在a+-1d    [-1,1]
    27%的数据会落在a+-2d    [-2,2]
    0.3%的数据会落在a+-3d    [-3,3]
    numpy.random.randn(d0, d1, ..., dn)
"""
import numpy as np

print(np.random.randn(10))
print(np.random.randn(3,2))
```
