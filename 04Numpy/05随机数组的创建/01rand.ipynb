```python
import numpy as np
# 设置随机数种子，种子值可以是任何整数，通常是正整数。
# np.random.seed(42)

"""
    rand: 返回一个或一组从0-1均匀分布的随机样本
    这个区间内的任意一个数值，被选中的概率是相等的
    numpy.random.rand(d0, d1, ..., dn)
    d0, d1, ..., dn： 这些参数定义了输出数组的形状
"""
print(np.random.rand())
print(np.random.rand(3))
print(np.random.rand(3,2))
```
