```python
"""
    randint:生成随机整数，指定范围内的均匀分布
    numpy.random.randint(low, high=None, size=None, dtype=int)
    low(包含)：随机数的起始点，如果提供了low而没有提供high，那么随机整数的范围从0-low（不包含low本身）
    high(不包含)：随机数的结束点

"""

import numpy as np

# 从0-10之间随机生成一个整数
print(np.random.randint(0,10))
```

```python

print(np.random.randint(0,10,size=(2,5),dtype=np.int16))
```
