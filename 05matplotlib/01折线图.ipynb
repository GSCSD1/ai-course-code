```python

import numpy as np
"""绘制正弦图线"""
# 创建x轴的数据
xs = np.arange(0,2*np.pi,0.1)
# 创建y轴的数据
ys = np.sin(xs)

# 打印x和y的值
print(xs)
print(ys)



```

```python
"""
    plt.plot:绘制线
    plot([x], y, [fmt], **kwargs)

    x ： 可以是一个列表或者数组，如果x未被指定，默认为rang(len(y))
    y : 可以是一个列表或者数组
    fmt: 定义图形的颜色和样式，格式化字符串
        颜色：r:红 g:绿 b:蓝 c:青 m:洋红 y:黄 k:黑 w:白
        标记样式：.:点 o:圆圈 s:方形 ^:三角形 D:菱形 x:叉号
        线型：- 实线(默认)    -- 虚线    : 点线    -.: 点划线
    **kwargs : 进一步定义图形的属性
        label ： 图例的标签
        linewidth: 线宽
        color: 线的颜色
        marker：标记样式
        markersize或ms: 标记点的大小
        markeredgecolor: 标记点边缘颜色
        markeredgewidth: 标记点边缘宽度
        markerfacecolor: 标记点内部颜色
        alpha : 图像的透明度
"""

from matplotlib import  pyplot as plt

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用雅黑字体
plt.rcParams['axes.unicode_minus'] = False  # 显示负号

# 绘制线
plt.plot(xs,ys,'ro-',
            label='正弦曲线',
            linewidth=3,
            markersize=5,
            markerfacecolor='y',
            alpha=1  # 透明度，0~1之间，越小越透明，越
         )

# 显示图例
plt.legend()
# 显示图
plt.show()
```
