```python
"""
    xlabel ： 设置x轴的标签
    ylabel ： 设置y轴的标签
    title :   设置图的标题  （显示在图标顶部的文本）
    subplot: 用于在一个图形窗口（figure）创建多个子图布局
"""

from matplotlib import  pyplot as plt
import numpy as np
# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用雅黑字体
plt.rcParams['axes.unicode_minus'] = False  # 显示负号

"""绘制正弦图线"""
# 创建x轴的数据
xs = np.arange(0,2*np.pi,0.1)
# 创建y轴的数据
ys = np.sin(xs)
# 绘制线
plt.plot(xs,ys,'ro-', label='正弦曲线', )
# 显示图例
plt.legend()
plt.xlabel("时间轴 t/s")
plt.ylabel("函数值 ")
plt.title("曲线图绘制")
# 显示图
plt.show()

```

```python

plt.subplot(1,2,1) # 创建2行1列的子图，并选择第一行第一列的子图
plt.plot(xs,ys)
plt.title("正弦曲线")

# 创建x轴的数据
xs = np.arange(0,2*np.pi,0.1)
# 创建y轴的数据
ys = np.cos(xs)

plt.subplot(1,2,2)  # 创建2行1列的子图，并选择第二行第一列的子图
plt.title("余弦曲线")
plt.plot(xs,ys)
# 用于自动调节图形窗口，使子图大小一致
plt.tight_layout()
```
