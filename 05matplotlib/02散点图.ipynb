```python
import numpy as np
"""绘制正弦图线"""
# 创建x轴的数据
xs = np.arange(0,2*np.pi,0.1)
# 创建y轴的数据
ys = np.sin(xs)

# 打印x和y的值
print(xs)
print(ys)
```

```python
"""

    ptl.scatter:
    scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)

    x，y : 数组或标量，代表的是散点的x和y坐标
    s ： 散点的大小，可以是标量或数组，用于指定每个点的大小
    c ： 散点的颜色 可以是单个颜色，也可以是数组，为每个点指定颜色，如果提供了cmap参数，则c可以是颜色的映射值
    cmap:colormap 用于将c中的数值映射到颜色
         viridis: 默认渐变 从暗紫色到亮黄色
         plasma: 从暗紫色到亮红色
         inferno:从黑色到亮黄色
         cividis：色盲友好的渐变
         coolwarm:蓝红色的渐变  适用于温度等数据的可视化
    marker：散点的标记样式  .:点 o:圆圈 s:方形 ^:三角形 D:菱形 x:叉号
    norm: 用于标准化c的颜色数据的normalize对象，确保数据在指定范围内的颜色映射
          none： 会根据c参数中传递的数值自动计算出最大值和最小值，并将这些颜色映射到[0,1]范围内
          normalize：线性映射
          LogNorm:   对数映射
          TwoSlopeNorm： 斜率中点映射
    vmin，vmax(可选) 分别是c中数据的最小值和最大值
    alpha：透明度
    linewidths：散点边缘的线宽
    edgecolors: 散点边缘的颜色
"""

from matplotlib import pyplot as plt

# 绘制散点图
plt.scatter(xs,ys,s=30,c=xs,cmap='viridis',edgecolors='r')

# 添加颜色条
plt.colorbar()

plt.show()
```

```python

# 生成100个均值为0，方差为1的正态分布随机数
ys = np.random.randn(100)
xs = np.linspace(0,10,100) # 生成x轴的值 100个点 生成线性间隔的数组 用于创建一个在指定范围内均匀分布的数值序列


# 绘制散点图
plt.scatter(xs,ys,s=30,c=ys)

# 添加颜色条
plt.colorbar()

plt.show()
```
