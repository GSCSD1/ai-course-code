```python
"""
    plt.pie: 用来绘制饼图

    pie(x, explode=None, labels=None, colors=None, autopct=None,
     shadow=False,  startangle=0, radius=1,
      wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False,
      normalize=True, hatch=None, data=None)

    x ： 饼图中每个扇形的尺寸   为非负的数组或序列 每个数加起来等于100
    explode: 一个与x相同长度的数组，用于指定每个扇形是否突出显示。如果explode[i]不为0，则会与其他扇形分离
    labels: 一个字符串列表，用于指定每个扇形的标签。
    colors:  一个颜色列表，用于指定每个扇形的颜色
    autopct: 格式化字符串 用来显示每个扇形的百分比 , 使用%.1f%% 来表示一个只保留一位小数的百分比
    shadow : 布尔值 用于指定是否为饼图添加阴影
    startangle: 饼图开始的角度，默认为0
    radius: 饼图的半径
    wedgeprops: 字典数据  用于指定每个扇形的属性，如边缘颜色或宽度
    textprops: 字典数据  用于指定饼图中标签的文本属性
    center: 一个元组  用于指定饼图的中心位置
    frame : 布尔值，用于指定是否为饼图添加一个坐标轴
    rotatelabels: 是否指定旋转标签以适应扇形
    normalize : 布尔值 如果为True，则x中的值将被归一化以使得它的总和为1
    hatch: 进行图案填充  :点 o:圆圈 s:方形 ^:三角形 D:菱形 x:叉号
"""
import numpy as np
from matplotlib import pyplot as plt
# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用雅黑字体
plt.rcParams['axes.unicode_minus'] = False  # 显示负号



labels = ["麻辣烫","鸡腿饭","牛肉面","汉堡","鸡架拌饭","冒菜"]
xs = np.array([18,17.3,15,13.5,16.5,19.5])

# 绘制饼图
plt.pie(xs,labels=labels,
        explode=[0.2,0,0,0,0,0],
        colors=["y","g","r","Purple","gray","#1F77B4"],
        autopct = "%.1f%%",
        shadow =  True,
        startangle = 0,
        radius = 1.5,
        textprops=dict(color="black"),
        frame=True,
        rotatelabels = False,
        hatch = None)
plt.legend()
plt.show()
```
