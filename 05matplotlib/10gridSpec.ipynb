```python
"""
    为子图提供更灵活的布局，在matplotlib中，默认的子图布局是通过plt.subplot或者
    plt.subplots等函数实现，但这些函数都是按照行列均匀划分子图，而GridSpec可以
    根据具体需求定制子图的位置和大小

    matplotlib.gridspec.GridSpec(nrows, ncols, figure=None, left=None,
    bottom=None, right=None, top=None, wspace=None, hspace=None,
    width_ratios=None, height_ratios=None)

    nrows：网格的行数
    ncols：网格的列数
    figure：可以指定需要绑定画布，没有提供将与当前活动的画布关联
    left： 子图与图形窗口左边缘之间的距离     范围是从0-1
    bottom： 子图与图形窗口下边缘之间的距离     范围是从0-1
    right： 子图与图形窗口右边缘之间的距离     范围是从0-1
    top： 子图与图形窗口上边缘之间的距离     范围是从0-1

    wspace：子图之间的水平间距
    hspace：子图之间的垂直间距
    width_ratios： 子图之间的宽度比例
    height_ratios： 子图之间的高度比例
"""

from matplotlib import pyplot as plt
from matplotlib.gridspec import  GridSpec

"""基本用法"""

fig = plt.figure()
# 创建四个网格
gs =  GridSpec(2,2)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

plt.show()
```

```python

"""
    子图跨越多行
"""
fig = plt.figure()
# 创建四个网格
gs =  GridSpec(3,3)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1:3,0])
ax4 = fig.add_subplot(gs[1,1:3])

plt.show()
```

```python
"""
    width_ratios： 子图之间的宽度比例
    height_ratios： 子图之间的高度比例
"""
fig = plt.figure()
# 创建四个网格
gs =  GridSpec(2,2,width_ratios=[1,3],height_ratios=[2,1])
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])
plt.show()
```
