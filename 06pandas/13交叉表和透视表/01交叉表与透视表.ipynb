```python

import pandas as pd

data = pd.read_csv('../../stock_day.csv')
data.head()
```

```python
data.index
```

```python
time =  pd.to_datetime(data.index)
```

```python
time.weekday
```

```python
data['week'] = time.weekday
data.head(10)
```

```python
import numpy as np

temp =  np.where(data['p_change']>0,1,0)
print(temp)
data['p_n'] = temp
```

```python
data.head(10)
```

```python
# 交叉表去求解在week中，涨跌的次数
count =  pd.crosstab(data['week'],data['p_n'])
print(count)

```

```python
# 获取第0列和第1列的总和
sum = count.sum(axis=1)
sum
```

```python
# 获得涨和跌的概率
ret =  count.div(sum,axis=0)
ret
```

```python
ret.plot(kind='bar',stacked=True)
```

```python
# 透视表求解在week中，涨跌的概率
data.pivot_table(['p_n'],index='week')
```
