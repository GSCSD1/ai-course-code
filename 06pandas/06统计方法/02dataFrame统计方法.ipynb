```python
"""
    计算dataframe中非nan值的数量
        DataFrame.count(axis=0, numeric_only=False)
    axis ： 0 则对每列进行计数
            1 则对行进行计数
    numeric_only：是否忽略非数值列
"""
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {
        'A':[1,2,np.nan,4],
        'B':[5,np.nan,np.nan,8],
        'C':['foo','bar','baz',np.nan]
    }
)

print(df)

# 计算每列中非nan值的数量
print(df.count(axis=0))

```

```python
# 计算每行中非nan值的数量
print(df.count(axis=1))
```

```python
"""
    计算所有值的和
    dataframe.sum(axis=0, skipna=True, numeric_only=None)
    axis: 沿着哪个轴方向进行计算
    skipna ； 跳过NaN值
    numeric_only: 为True则只对数字类型进行运行
"""
print(df)

# 求每列的总和
print(df.sum(axis=0,numeric_only=True))
```

```python
# 求每行的总和
print(df.sum(axis=1,numeric_only=True))
```

```python
"""
    求平均值
    dataframe.mean(axis=0, skipna=True, numeric_only=None)
    axis: 沿着哪个轴方向进行计算
    skipna ； 跳过NaN值
    numeric_only: 为True则只对数字类型进行运行

"""
print(df)
# 求每列的平均值
print(df.mean(axis=0,numeric_only=True))
```

```python
"""
    中位数
    dataframe.median(axis=0, skipna=True, numeric_only=False)

    axis: 沿着哪个轴方向进行计算
    skipna ； 跳过NaN值
    numeric_only: 为True则只对数字类型进行运行
"""
print(df)
# 求每列的 中位数
print(df.median(axis=0,numeric_only=True))
```

```python
"""
    dataframe.min(axis=0, skipna=True, numeric_only=False)
    dataframe.max(axis=0, skipna=True, numeric_only=False)
"""
print(df)
# 求每列的 最小值和最大值
print(df.min(axis=0,numeric_only=True))
print(df.max(axis=0,numeric_only=True))
```

```python
"""
    方差
    dataframe.var(axis=None, skipna=True, ddof=1, numeric_only=False)

    ddof： Delta Degress of freedom(自由度调整) 用于贝塞尔矫正  以得到样本方差的估计
    对于无偏估计（样本方差） ddof通常设置为1 如果计算总体方差 应该把ddof设置为0

    样本方差：从样本中估计估计总体的误差
    总体方差： 计算总体的误差
"""
print(df)
# 求每列的方差
print(df.var(axis=0,numeric_only=True))
```

```python
"""
    标准差
    dataframe.std(axis=None, skipna=True, ddof=1, numeric_only=False,**kwargs)
"""
print(df)
# 求每列的标准差
print(df.std(axis=0,numeric_only=True))
```
