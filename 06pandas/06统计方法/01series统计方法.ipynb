```python

"""
    count : 计算非缺失值（nan  None）的个数
"""

import numpy as np
import pandas as pd
s1 = pd.Series([1,5,3,np.nan,None,13,76])
print(s1.count())
```

```python
"""
    计算所有值的和
    Series.sum(axis=None, skipna=True, numeric_only=None, min_count=0)
    skipna ； 跳过NaN值
    numeric_only: 为True则只对数字类型进行运行   只针对dataframe
    min_count：计算总和之前至少需要多少个非NaN值，如果小于则为NaN
"""

s1 = pd.Series([1,5,3,10,20,np.nan])

print(s1.sum(skipna=False))
```

```python
"""
    求平均值
    Series.mean(axis=None, skipna=True, numeric_only=None)
"""

s1 = pd.Series([1,5,3,10,20])
print("平均值：",s1.mean(skipna=False))
```

```python
"""
    中位数
    Series.median(axis=0, skipna=True, numeric_only=False)
"""

print("中位数：",s1.median(skipna=False))
```

```python
"""
    Series.min(axis=0, skipna=True, numeric_only=False)
    Series.max(axis=0, skipna=True, numeric_only=False)
"""
print("最小值：",s1.min(skipna=False))
print("最大值：",s1.max(skipna=False))
```

```python
"""
    方差
    Series.var(axis=None, skipna=True, ddof=1, numeric_only=False)

    ddof： Delta Degress of freedom(自由度调整) 用于贝塞尔矫正  以得到样本方差的估计
    对于无偏估计（样本方差） ddof通常设置为1 如果计算总体方差 应该把ddof设置为0

    样本方差：从样本中估计总体的误差
    总体方差： 计算总体的误差
"""
s1 = pd.Series([1,5,3,10,20])
print("样本方差：",s1.var(skipna=False))
```

```python
"""
    Series.std(axis=None, skipna=True, ddof=1, numeric_only=False,**kwargs)
"""
print("样本标准差：",s1.std(skipna=False))
```
