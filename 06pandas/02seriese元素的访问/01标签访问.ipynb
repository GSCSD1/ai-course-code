```python
"""
    标签访问（单个值）
"""
import pandas as pd
series1 = pd.Series([12,64,75,77,17],index=['a','b','c','d','e'])

print(series1[1])  # 下标访问
print(series1['b'])  # 标签访问
```

```python

"""
    函数实现下标（位置）访问和标签访问
"""
print(series1.iat[1])  # 下标/位置访问
print(series1.at[1])  # 标签访问
```

```python
"""    注意事项：    当位置与标签产生冲突时 ： 优先级 ： 标签索引大于位置索引"""

import pandas as pd
series1 = pd.Series([12,64,75,77,17],index=[2,3,4,1,5])
print(series1[1])
```
