```python


import pandas as pd
series1 = pd.Series([12,64,75,77,17],index=['a','b','c','d','e'])

# 位置切片
print(series1[:4]) # 取前4个
print(series1[::2]) # 按偶数次取
```

```python
# 标签切片     起始值和结束值都包含在内
print(series1['a':'d']) # 取前4个
print(series1['a':'e':2]) # 按偶数次取
```

```python
"""
    函数实现：
            loc : 标签切片
            iloc : 位置切片

"""
# 位置切片
print(series1.iloc[:4]) # 取前4个
# 标签切片
print(series1.loc['a':'d']) # 取前4个

```

```python
"""    注意事项：    当位置与标签产生冲突时 ： 优先级 ： 位置切片大于标签切片"""

import pandas as pd
series1 = pd.Series([12,64,75,77,17],index=[2,3,4,1,5])
print(series1[2:4])
```
