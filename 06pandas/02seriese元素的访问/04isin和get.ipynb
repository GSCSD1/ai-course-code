```python
"""
    isin:用于判断Series中的每个元素是否在指定一组值中
    series.isin(values)
    values : 可迭代对象，用于指定要进行判断的一组值
    返回一个与原series 长度相同的布尔型Series ，其中对应的值为True表示该位置的元素在指定的值集合
    中,false则表示不在
"""
import pandas as pd
# 创建series对象
data = [1,2,3,4,5,6]
series = pd.Series(data,index=['a','b','c','d','e','f'])
# 判断一组值是否在series中,并把满足条件的位置索引返回出来
print(series.isin([3,1]))
```

```python
"""
    通过标签来获取series中的元素
    pd.Series.get(key,default=None)
    Key: 你想要获取元素的标签
    default：可选参数，如果key不在标签中，返回这个默认值，如果没有指定，默认为none
"""
import pandas as pd

s = pd.Series(['apple','banana','cherry'],index=[2,3,1])

#获取标签为2的值
print(s.get(2))
# 标签不存在是，返回提示词
print(s.get(5,'Not Found'))

```
