```python
"""
    isnull()  判断哪个值是缺失值，如果是缺失值则返回True，如果不是则会返回False
"""
df = pd.DataFrame([[1,2,np.nan],[4,np.nan,6],[5,6,7]])
df
```

```python
print("获取一维的缺失值")
df = pd.Series([1,2,np.nan,4,np.nan,6,5,6,7])
df[df.isnull()]
```

```python
df = pd.DataFrame([[1,2,np.nan],[4,np.nan,6],[5,6,7]])
print('isnull 方法判断是否是缺失值')
df.isnull()
```

```python
print(df.isnull())
```

```python
np.where(df.isnull())
```

```python
print("获取一维的非空缺失值")
df = pd.Series([1,2,np.nan,4,np.nan,6,5,6,7])
df[df.notnull()]
```
