```python


"""
    class pandas.Series(data=None, index=None, dtype=None, name=None,copy=None, fastpath=False)

    data :数据  可以为标量、列表、元组、字典、数组
    index：数组或列表，用于定义Series的索引，如果未提供，则默认从0开始
    dtype：数组的数据类型
    name：Series标识名，用于后续的索引和操作
    copy： 布尔值，默认为False。如果为True，则复制数据，如果为false，则尽可能避免复制数据，仅影响ndarray输入
    fastpath：布尔值，通常不用用户指定。内部优化标志，True：运行Series构造函数绕过一些检查和验证步骤，加快series的创建速度
"""
import pandas as pd
import numpy as np

"""
    标量方式创建1维数组
"""
Series =  pd.Series(20)
print(Series)
print(type(Series))
```

```python
"""
    列表方式创建1维数组
"""
Series =  pd.Series(np.random.rand(6),index=['a','b','c','d','e','f'],dtype=np.float64,name="合格率")
print(Series)
```

```python
"""
    字典方式创建1维数组
    字典的键是索引 字典的值就是该索引对应的值。

    注意： 如果用字典创建了series，并且指定了与字典的键不同的index，则生成的数据就是NAN

    Pandas: nan(not a number) 是一个特殊的浮点数，用于表示缺失数据或无效数据
"""

Series =  pd.Series({"语文":70,"数学":87,"英语":90},index=['A','B','C'])
print(Series)
```
