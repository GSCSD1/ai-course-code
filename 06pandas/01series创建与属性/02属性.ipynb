```python
import pandas as pd
import numpy as np
""" index和values和name属性"""
Series =  pd.Series(np.random.rand(6),index=['a','b','c','d','e','f'])
print( Series)
# 索引
print("行索引：\n",Series.index)
# 值
print("值：\n",Series.values)
# 名称
print("名称：\n",Series.name)
```

```python
"""dtype  shape   size"""
print(Series.dtype)
print(Series.shape)
print(Series.size)
```

```python
"""empty"""
print(Series.empty)
```

```python
"""hasnans  返回数组中是否包含NaN值"""
Series =  pd.Series([1,5,14,np.nan],index=['a','b','c','d'])
print(Series)
print(Series.hasnans)
```

```python
"""is_unique  判断数组中有没有重复元素 """
Series =  pd.Series([1,5,1,np.nan],index=['a','b','c','d'])
print(Series.is_unique)
```

```python
"""nbytes获取数组元素所占用的内存大小 """
print(Series.nbytes)
```

```python
"""axes:  用于返回series对象行轴标签的列表。"""

import pandas as pd
# 创建Series数组
series = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd','e'], dtype='int64')
# 获取数组的行轴标签
print(series.axes)
```

```python
"""array  返回Series的底层数组，包括数组的元素、数组的长度及数组元素的数据类型"""
print(series.array)
```

```python
"""attrs 属性"""
series = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd','e'], dtype='int64')
series.attrs = {'name': 'series', 'type': 'numeric','time': '2025-08-13T01:31:03.898706Z'}

print(series.attrs)
```

```python
"""
     is_monotonic_decreasing  判断是否为降序
     is_monotonic_increasing  判断是否为升序
"""

print(series.is_monotonic_decreasing)
print(series.is_monotonic_increasing)
```
