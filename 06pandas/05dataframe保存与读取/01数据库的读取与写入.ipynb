```python


import sqlite3
import pandas as pd

# 数据库连接
conn =  sqlite3.connect("../../../test.db")

# 读取表
df = pd.read_sql('select * from order_items',con = conn)

display(df)

display(df[['order_id','quantity','price']])

print("---------------------")
```
