```python
"""
    index: 返回行索引
    columns : 返回列索引
    values ： 返回数据
    dtypes：返回每一列的数据类型
    shape : 返回表格的形状
    size ： 元素个数
    empty: 判断是否为空
    T ： 转置
    ndim：维度  为2
    attrs： 存储元数据
"""

import pandas as pd
df =  pd.DataFrame([['小明',90],['小王',80],['小李',70]],columns=['姓名','成绩'])
print(df)


print("行索引：\n",df.index)
print("列索引：\n",df.columns)
print("数据：\n",df.values)
print("数据类型：\n",df.dtypes)
print("表格的形状：\n",df.shape)
print("元素的个数：\n",df.size)
print("是否为空：\n",df.empty)
print("转置：\n",df.T)
print("维度：\n",df.ndim)

df.attrs = {'name': 'dataFrame', 'type': 'numeric','time': '2025-08-13T01:31:03.898706Z'}
print("元数据：\n",df.attrs)
```
