```python
"""
    用来绘制dataframe数据的可视化图表
    dataframe.plot(*args, **kwargs)

    kind：
            ‘line’ 'bar'
            'barh' 水平柱状图  ‘hist’ 直方图  ‘box’ 箱线图 ‘kde’ 核密度估计图 ‘area’ 面积图
            'pie' 'scatter'
            'hexbin' 六边形箱图
    ax：用于在指定轴上绘制图表
    figsize:图表的尺寸
    use_index:是否使用Series的索引作为x轴标签  默认True
    title：标题
    gird：是否显示网格线
    legend：是否显示图例
    xticks：x轴的刻度位置
    yticks：y轴的刻度位置
    xlim：x轴的范围
    ylim：y轴的范围
    color：绘制颜色
    label：图例的标签
"""
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'  # 设置字体为 SimHei（黑体）
data={
    '编号':[1,2,3],
    '姓名':['小明','小刚','小红'],
    '年龄':[20,21,19],
    '成绩':[78,64,89]
}
df = pd.DataFrame(data)
df
```

```python
from matplotlib import pyplot as plt

df.plot(kind='bar')
plt.show()
```
