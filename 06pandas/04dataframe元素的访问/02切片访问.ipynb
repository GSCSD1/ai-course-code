```python
import pandas as pd
"""列表创建"""
df =  pd.DataFrame([['小明',90,True],['小王',80,False],['小李',70,True]],columns=['姓名','成绩','到课率'],index=['A','B','C'])
display("数据集:",df)

#   不允许符号方式进行切片
print("取出姓名和成绩列数据:\n",df[['姓名','成绩']])
print("取出姓名和成绩列数据:\n",df.loc['A':'C','姓名':'成绩'])    # 标签切片
print("取出A行中第0列和第1列数据\n",df.iloc[0:1,0:2])    # 位置切片
```
