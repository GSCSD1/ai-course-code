```python
"""
    查看开头和末尾元素
    DataFrame.head(n=5)   默认返回5行
    DataFrame.tail(n=5)   默认返回最后5行
"""
import pandas as pd
data={
    '编号':[1,2,3,4,5,6,7,8,9],
    '姓名':['小明','小刚','小红','小王','小李','小张','小赵','小钱','小孙'],
    '年龄':[20,21,19,18,17,16,15,14,13],
    '成绩':[78,64,89,95,100,98,96,97,99],
}

df = pd.DataFrame(data)
print(df)
print(df.head(3))
print(df.tail(3))
```
