```python

import pandas as pd

"""列表创建"""
df =  pd.DataFrame([['小明',90,True],['小王',80,False],['小李',70,True]],columns=['姓名','成绩','到课率'],index=['A','B','C'])
print(df)

# print(df[0][1]) # 不存在位置访问
print(df['成绩']['A'])
"""取单列数据"""
print(df['姓名'])
```

```python
"""取多列数据"""
print(df[['成绩','到课率']])
```

```python

"""
    用函数实现当个值访问
         标签访问   at[行标签,列标签]
         位置访问   iat
"""
df =  pd.DataFrame([['小明',90,True],['小王',80,False],['小李',70,True]],columns=['姓名','成绩','到课率'],index=['A','B','C'])
print(df)
print(df.at['A','成绩'])

print(df.iat[0,1])
```
